```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/Dress+Sales.csv")

# Drop 'Dress_ID' column for analysis
sales_data = df.drop(columns=['Dress_ID'])

# Convert column names to datetime
sales_data.columns = pd.to_datetime(sales_data.columns, dayfirst=True)

# Transpose the DataFrame so dates become rows
sales_transposed = sales_data.T

# Sum across all dresses for each date (now a row)
sales_transposed['TotalSales'] = sales_transposed.sum(axis=1)

# Convert the index (which are dates) to a Series so we can extract months
sales_transposed['Month'] = sales_transposed.index.month

# Define season mapping function
def month_to_season(month):
    if month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    elif month in [12, 1, 2]:
        return 'Winter'
    else:
        return 'Spring'

# Map each row (date) to a season
sales_transposed['Season'] = sales_transposed['Month'].map(month_to_season)

# Group by season and sum total sales
seasonal_sales = sales_transposed.groupby('Season')['TotalSales'].sum()

# Find lowest season and its value
lowest_season = seasonal_sales.idxmin()
lowest_value = seasonal_sales.min()

print(f"Season with lowest sales: {lowest_season}")
print(f"Total sales in that season: {lowest_value:.2f}")

```
