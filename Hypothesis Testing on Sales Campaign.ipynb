```python
import numpy as np
from scipy import stats

#Revenue data
# Group A: 10 dates revenue in summer without campaign -Control group(no campaign)
group_a = np.array([120, 112, 118, 130, 125, 128, 115, 121, 123, 119])

# Group B: 10 days revenue in summer same period with campaign - Test group (with campaign)
group_b = np.array([115, 120, 112, 128, 122, 125, 110, 117, 125, 150])

# Hypotheses:
# H0 (null): There is no difference in average revenue between groups
# H1 (alt): There is a difference in average revenue between groups

# Perform two-sample (independent) t-test
t_statistic, p_value = stats.ttest_ind(group_b, group_a, equal_var=False)  # Welch's t-test (safer)

# Significance level
alpha = 0.05

# Decision logic
if p_value < alpha:
    decision = "Reject Null Hypothesis"
    conclusion = "Statistically significant difference: The campaign likely had an impact."
else:
    decision = "Fail to Reject Null Hypothesis"
    conclusion = "No significant difference: The campaign may not have had a meaningful impact."

# Output
print("Group A (Without campaign) Mean Revenue:", np.mean(group_a))
print("Group B (With Campaign) Mean Revenue:", np.mean(group_b))
print("T-statistic:", round(t_statistic, 4))
print("P-value:", round(p_value, 4))
print("Decision:", decision)
print("Conclusion:", conclusion)
```
